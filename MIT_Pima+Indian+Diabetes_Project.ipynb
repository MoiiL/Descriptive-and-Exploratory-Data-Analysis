#### Answer : 
The Heatmap identifies the correlation between the various variables in the dataset and highlights the intensity of correlation using colours that range from bright colours (indicating positive correlation) to darker colours (indicating the nature of negative correlation). Some of the observations incurred are as follows:
1. Relatively Strong positive correlation: 
   The variables Age and Pregnancies have a strong positive correlation of 0.54 indicating that fertility peaks among the younger women group.
2. Moderately strong positive correlation:
   The variables Insulin - Glucose, BMI - Glucose, Age - Glucose, SkinThickness- BloodPressure, BMI - BloodPressure, Age - BloodPressure, Insulin - SkinThickness, BMI - SkinThickness and BMI - Insulin have a moderately positive correlation ranging from 0.2 - 0.44.
